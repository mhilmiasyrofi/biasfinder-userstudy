```python
def is_placeholder(token):
    return token[0] == "<" and token[-1] == ">" 

def is_punctuation(token):
    return token == "'s" or token[0] in string.punctuation

left_mark = "<start>"
right_mark = "<end>"
        
i = 0
j = 0
k = 0

id = 0
original = df.iloc[id]["original_text"].split()
template = df.iloc[id]["template"].split()
mutant = df.iloc[id]["mutated_text"].split()

fmt_original = []
fmt_mutant = []

# print(template)
# print(original)

while j < len(template) :
    if is_placeholder(template[j]) :
        fmt_original.append(left_mark)
        fmt_mutant.append(left_mark)
        
        fmt_original.append(original[i])
        fmt_mutant.append(mutant[k])
        i += 1
        j += 1
        k += 1
        next_token = template[j]
        while is_punctuation(next_token):
            j+=1
            next_token = template[j]            

        while i < len(original) and original[i] !=  next_token :
            fmt_original.append(original[i])
            i += 1
        while k < len(mutant) and mutant[k] !=  next_token :
            fmt_mutant.append(mutant[k])
            k += 1
        fmt_original.append(right_mark)
        fmt_mutant.append(right_mark)
    else :
        fmt_original.append(original[i])
        fmt_mutant.append(mutant[k])
        i += 1
        j += 1
        k += 1
        
# print(fmt_original)
# print(fmt_mutant)
```

```python
def format_text(text, template):
    left_mark = "<start>"
    right_mark = "<end>"

    i = 0
    j = 0

    text = text.split()
    template = template.split()
    
    print(text)
    print(template)

    fmt_text = []

    while j < len(template) :
        if is_placeholder(template[j]) :
            fmt_text.append(left_mark)

            fmt_text.append(text[i])
            i += 1
            j += 1
            if j < len(template) :
                next_token = template[j]
            
                while j < len(template) and (is_punctuation(next_token) or is_placeholder(next_token)):
                    j += 1
                    if j < len(template) : 
                        next_token = template[j]

                while i < len(text) and text[i] !=  next_token :
                    fmt_text.append(text[i])
                    i += 1                
                fmt_text.append(right_mark)
            else :
                while i < len(text):
                    fmt_text.append(text[i])
                    i += 1                
                fmt_text.append(right_mark)
                
        else :
            if i < len(text) :
                print("i",i)
                print("text", text[i])
                fmt_text.append(text[i])
                i += 1
            j += 1
            
    return " ".join(fmt_text)


id = 278
format_text(df.iloc[id]["original_text"], df.iloc[id]["template"])
```
